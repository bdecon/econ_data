Identify how many own kids each person has in their household, first under 18, then under age 5. 

```python
# Import libraries
import os
import numpy as np
import pandas as pd

os.chdir('/home/brian/Documents/CPS/data/clean/')
```

```python
columns = ['MONTH', 'PARENT', 'SPOUSE',
           'PRNMCHLD', 'PULINENO', 
           'STATE', 'AGE',
           'HHID', 'HHID2']
df = (pd.read_feather('cps2019.ft', columns=columns)
        .query('MONTH == 3')
        .astype({'PULINENO': 'int8'}))

kids = (df.query('AGE < 18 and PARENT > 0')
          .groupby(['HHID', 'HHID2', 'STATE', 'PARENT']).PULINENO.count()
          .reset_index()
          .rename({'PARENT': 'PULINENO', 
                   'PULINENO': 'NUMCHILD1'}, axis=1))
result = (df.merge(kids, 'left')
            .merge(kids.rename({'PULINENO': 'SPOUSE', 
                                'NUMCHILD1': 'NUMCHILD2'}, axis=1), 'left'))
df['NCHILDU18'] = (np.where(result['NUMCHILD1'] > 0, result['NUMCHILD1'],
                   np.where(result['NUMCHILD2'] > 0, result['NUMCHILD2'], 0))
                     .astype('int8'))

data = (df.query('AGE < 5 and PARENT > 0')
          .groupby(['HHID', 'HHID2', 'STATE', 'PARENT']).PULINENO.count()
          .reset_index()
          .rename({'PARENT': 'PULINENO', 
                   'PULINENO': 'NUMCHILD1'}, axis=1))
result2 = (df.merge(data, 'left')
             .merge(data.rename({'PULINENO': 'SPOUSE', 
                                 'NUMCHILD1': 'NUMCHILD2'}, axis=1), 'left'))
df['NCHILDU5'] = (np.where(result2['NUMCHILD1'] > 0, result2['NUMCHILD1'],
                  np.where(result2['NUMCHILD2'] > 0, result2['NUMCHILD2'], 0))
                    .astype('int8'))

df = df.drop(['NUMCHILD1', 'NUMCHILD2', 'PARENT', 'SPOUSE'], axis=1)
```
