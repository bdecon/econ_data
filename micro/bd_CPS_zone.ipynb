System for splitting CPS into 70 geographic zones with economic and statistical significance. 

```python
import pandas as pd
import numpy as np
from pathlib import Path

cps_dir = Path('/home/brian/Documents/CPS/data/clean/')
```

```python
cols = ['YEAR', 'MONTH', 'ZONEERA', 'ZONE', 'PWSSWGT']
for year in list(range(1995, 2026)):
    df = (pd.read_feather(cps_dir / f'cps{year}.ft', columns=cols))
    empty = df.query('ZONE != ZONE', engine='python')
    print(f'{year}: {len(empty)}')
```

```python
cols = ['YEAR', 'MONTH', 'ZONEERA', 'ZONE', 'PWSSWGT', 'TLWK']
for year in list(range(2022, 2026)):
    df = (pd.read_feather(cps_dir / f'cps{year}.ft', columns=cols))
    for month in np.sort(df.MONTH.unique()):
        mdata = df.query('MONTH == @month')
        tlwk = np.average(mdata['TLWK'].fillna(0), weights=mdata['PWSSWGT'])
        print(f'{year}-{month:02}-01: {tlwk * 100:.1f}')
```
